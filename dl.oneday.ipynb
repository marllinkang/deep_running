```python
import os
path = "../EDA/data/deep_running_data/dl.oneday.project o,x/"
os.listdir(path)
```

```python
O_X_names = os.listdir(path)
O_X_labels = [each.split('_')[0] for each in O_X_names]
O_X_file_id =[each.split('.')[1] for each in O_X_names]
```

```python
from tqdm.notebook import tqdm
from skimage.color import rgb2gray
import numpy as np

images = []
bar_total = tqdm(O_X_names)
for file in bar_total:
    image = mpimg.imread(path+file)
    images.append(resize(image, (300, 300, 3)))

images = np.array(images)
```

```python
images.shape, O_X_labels[:5]
```

```python
images[:3]
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder.fit(O_X_labels)
O_X_labels_encoded = encoder.transform(O_X_labels)
O_X_labels_encoded[:3], encoder.classes_
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train , y_test =\
                train_test_split(images, O_X_labels_encoded, test_size =0.2,
                                 random_state = 13, stratify  = O_X_labels_encoded)
X_train.shape, X_test.shape
```

```python
samples = random.choices(population = range(0, 244), k =8 )

plt.figure(figsize =(14,12))
for idx, n in enumerate(samples):
    plt.subplot(2, 4, idx+1)
    plt.imshow(X_train[n], cmap ='Greys',interpolation = 'nearest')
    plt.title(y_train[n])

plt.tight_layout()
plt.show()
```

```python
from tensorflow.keras import layers, models

model = models.Sequential([layers.Conv2D(32,(3,3), activation ='relu', input_shape=(300,300,3)),
                           layers.MaxPooling2D(pool_size=(2,2), strides =(2,2)),
                           layers.Dropout(0.25),
                           
                           layers.Conv2D(64,(3,3), activation='relu',padding ='same'),
                           layers.MaxPooling2D(pool_size=(2,2)),
                           layers.Dropout(0.25),

                           layers.Conv2D(64,(3,3), activation='relu',padding ='same'),
                           layers.MaxPooling2D(pool_size=(2,2)),
                           layers.Dropout(0.25),
                           
                           layers.Flatten(),
                           layers.Dense(512, activation ='relu'),
                           layers.Dropout(0.25),
                           layers.Dense(2, activation ='softmax')])



model.summary()
```

  ~~

```python
model.compile(optimizer='adam', loss = 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
import time

start_time = time.time()
hist = model.fit(X_train.reshape(244, 300, 300, 3), y_train,
                 epochs = 35, verbose =1, # 5
                 validation_data = (X_test.reshape(62, 300, 300, 3), y_test))
print("fit time :", time.time() - start_time)
```

 5      10   ,  35     50  .

```python
score = model.evaluate(X_test, y_test)
print('Test loss :', score[0])
print('Test accuracy :', score[1])
```

```python
plot_target = ['loss', 'val_loss', 'accuracy', 'val_accuracy']
plt.figure(figsize=(12,8))

for each in plot_target:
    plt.plot(hist.history[each], label = each)
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
predicted_result = model.predict(X_test)
predicted_labels = np.argmax(predicted_result, axis=1)
predicted_labels[:10]
```

```python
wrong_result =[]

for n in range(0, len(y_test)):
    if predicted_labels[n] != y_test[n]:
        wrong_result.append(n)
len(wrong_result)
```

```python
plt.figure(figsize =(14,12))

for idx, n in enumerate(wrong_result):
    plt.subplot(1, 4, idx +1)
    plt.imshow(X_test[n], cmap='Greys', interpolation='nearest')
    plt.title('Label :' + str(y_test[n])+ ' Predict :' + str(predicted_labels[n]))
    plt.axis('off')

plt.show()
    
```
